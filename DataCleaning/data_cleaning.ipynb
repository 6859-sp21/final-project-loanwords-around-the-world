# Import Base Data

## Forms

```python
forms_df = pd.read_csv('wold_dataset_original/forms.csv')
forms_df
```

```python
print(forms_df.columns.tolist())

to_drop = []

for column in forms_df.columns:
    na_count = forms_df[column].isna().sum()
    print(f'{column} NA count: {na_count}')
    if na_count > 50000 and column != 'original_script':
        to_drop.append(column)
        
forms_df = forms_df.drop(columns = to_drop)
forms_df.head()


print((forms_df.Source == forms_df.Language_ID).value_counts())

forms_df = forms_df.drop(columns = ['Source', 'Contribution_ID'])

forms_df.head()
```

# Data Cleaning

```python
full_df = unit_df.merge(word_df, left_on = 'pk', right_on = 'pk').dropna(subset = ['language_pk'])

word_data_columns = full_df.columns.tolist()[1:]

print(word_data_columns)

full_df
```

```python
full_borrow_df = loan_df.merge(full_df.rename(mapper = lambda x: x + '_source', axis = 1), 
                               left_on = 'source_word_pk', right_on = 'pk_source', 
                               suffixes = ['', '_source']).drop(columns = ['jsondata_source', 'pk_source', 'age_score_source', 
                                                                           'borrowed_score_source', 'borrowed_source', 
                                                                           'simplicity_score_source', 'analyzability_source', 
                                                                           'polymorphic_type_source',
                                                                           'markup_description_source'])
# full_borrow_df.columns = [x + '_source' if x in word_data_columns else x for x in full_borrow_df.columns]

full_borrow_df = full_borrow_df.merge(full_df.rename(mapper = lambda x: x + '_target', axis = 1),
                                      left_on = 'target_word_pk', right_on = 'pk_target', 
                                      suffixes = ['', '_target']).drop(columns = ['polymorphic_type_target',
                                                                                  'markup_description_target'])

full_borrow_df.columns = [x + '_target' if x in word_data_columns else x for x in full_borrow_df.columns]

print(full_borrow_df.columns.tolist())

full_borrow_df = full_borrow_df[full_borrow_df.age_score_target.notna()]

full_borrow_df['first_attested'] = full_borrow_df.age_score_target.map({1.0:int(1000), 0.9:int(1500), 0.8:int(1800), 0.7:int(1900), 0.6:int(1950), 0.5:int(2007)})

full_borrow_df = full_borrow_df.merge(language_df.rename(mapper = lambda x: 'language_' + x + '_source', axis = 1), 
                                      left_on = 'language_pk_source', right_on = 'language_pk_source')

full_borrow_df = full_borrow_df.merge(language_df.rename(mapper = lambda x: 'language_' + x + '_target', axis = 1), 
                                      left_on = 'language_pk_target', right_on = 'language_pk_target')

full_borrow_df
```
